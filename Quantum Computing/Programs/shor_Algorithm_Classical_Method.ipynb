```python
import math
import matplotlib.pyplot as mpl
```

```python
N = 400
a = 13
```

```python
z = list(range(N))
y = [((a ** z_i) % N) for z_i in z]

mpl.plot(z, y)
mpl.xlabel('z')
mpl.ylabel(f"{a} ^ z mod {N}")
mpl.show()
```

```python
r = 0
while(a < N):
    try:
        r = z[y[1:].index(1) + 1]
    except ValueError as v:
        print(v)
        a = a + 1
        y = [((a ** z_i) % N) for z_i in z]
    else:
        break
```

```python
if (r % 2 == 0):
    x = (a ** (r//2)) % N
    print(f"x = {x}")
    if((x + 1) % N ):
        print(f"{math.gcd(N, x + 1)}, {math.gcd(N, x - 1)}")
    else:
        print(f"{x} + 1 = 0 (mod {N})")
else:
    print(f"r = {r} is odd.")
```
