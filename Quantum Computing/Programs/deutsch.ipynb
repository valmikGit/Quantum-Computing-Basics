```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
from qiskit_ibm_provider import *
%matplotlib inline
```

```python
my_Circuit = QuantumCircuit(2, 1)
```

```python
my_Circuit.h(qubit=0)
my_Circuit.x(qubit=1)
my_Circuit.h(qubit=1)

my_Circuit.barrier()

# Here our function f is the CNOT operation. It is a balanced function. Thus, we must get 1 as a output with high probability.
my_Circuit.cnot(control_qubit=0, target_qubit=1)

my_Circuit.barrier()

my_Circuit.h(0)

my_Circuit.barrier()

my_Circuit.measure([0], [0])
my_Circuit.draw(output='mpl')
```

```python
qasm_Simulator = Aer.get_backend('qasm_simulator')
my_Counts = execute(my_Circuit, backend=qasm_Simulator, shots=1024).result().get_counts()
plot_histogram(my_Counts)
```
